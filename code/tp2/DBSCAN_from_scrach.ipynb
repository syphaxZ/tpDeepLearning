```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris, load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Définition de la classe DBSCAN et des fonctions nécessaires

```python
class DBSCAN:
    def __init__(self, eps=0.5, min_samples=5):
        self.eps = eps
        self.min_samples = min_samples
        self.labels_ = None    
        
    def fit(self, X):
        n_samples = X.shape[0]
        self.labels_ = [None] * n_samples  # Initialisation des labels à None (non visité)
        cluster_id = 0                     # Initialisation du label du premier cluster
        for i in range(n_samples):
            if self.labels_[i] == None:  # Si le point n'a pas encore été visité
                neighbors = self.region_query(X, i) # Récupérer se voisins
                if len(neighbors) < self.min_samples: # Si le nombre de voisins n'est pas satisfait
                    self.labels_[i] = -1  # Marquer comme bruit
                else:
                    self.expand_cluster(X, self.labels_, i, neighbors, cluster_id)  # Etendre le cluster   
                    cluster_id += 1 # Créer un nouveau cluster
    # Fonction de recherche des voisins
    def region_query(self, X, point_idx):
        neighbors = []
        for i in range(len(X)):
            if np.linalg.norm(X[point_idx] - X[i]) < self.eps:
                neighbors.append(i)
        return neighbors                                

    
    # Fonction d'expansion d'un cluster
    def expand_cluster(self, X, labels, point_idx, neighbors, cluster_id):
        labels[point_idx] = cluster_id
        i = 0
        while i < len(neighbors):
            neighbor_idx = neighbors[i]            
            if labels[neighbor_idx] == -1:
                labels[neighbor_idx] = cluster_id            
            if labels[neighbor_idx] == None:
                labels[neighbor_idx] = cluster_id
                new_neighbors = self.region_query(X, neighbor_idx)
                if len(new_neighbors) >= self.min_samples:
                    neighbors.extend(new_neighbors)            
            i += 1
```

# Chargement des données

```python
# Charger les jeux de données Iris et Wine depuis sklearn.datasets
datasets = {
    'Iris': load_iris().data,
    'Wine': load_wine().data,
}

# Charger le jeu de données Wholesale Customers depuis un fichier CSV
wholesale_data = pd.read_csv('../../data/Wholesale customers data.csv')


# Ajout de 'wholesale_data' au dictionnaire datasets
datasets['Wholesale'] = wholesale_data.values
```

# Appliquer DBSCAN from scratch sur chaque jeu de données

```python
# Initialisation du scaler pour la Normalisation
scaler = StandardScaler()

results = {} # dictionnaire pour stocker les labels
for name, X in datasets.items():
    
    X_scaled = scaler.fit_transform(X)  # Normalisation
    
    # Réduction de dimension à 2 avec une PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)
    
    # Application de DBSCAN from scratch
    dbscan_from_scratch = DBSCAN(eps=0.5, min_samples=5)
    dbscan_from_scratch.fit(X_pca) 
    
    # Stocker les labels de DBSCAN from scratch
    results[name] = {'dbscan_from_scratch_labels': dbscan_from_scratch.labels_}
    
    # Visualisation
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_from_scratch.labels_, cmap='viridis')
    plt.title(f"DBSCAN from scratch sur les données {name}")
    
    # Ajouter une légende
    handles, labels = scatter.legend_elements()
    
    
    plt.legend(handles, labels, title='Labels')
    plt.show()
```
