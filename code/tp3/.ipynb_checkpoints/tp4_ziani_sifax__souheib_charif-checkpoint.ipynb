```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# Affichage des 10 premières images
import matplotlib.pyplot as plt
fig = plt.figure()
for i, digit in enumerate(digits.images[:10]):
 fig.add_subplot(1,10,i+1)
 plt.imshow(digit)
plt.show()
```

```python
import numpy as np
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
X, y = digits.data, digits.target
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
accuracy = clf.score(X,y)
print(accuracy)
```

```python
from sklearn.model_selection import train_test_split
# 90% des données pour le test, 10% pour l'apprentissage
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.90)
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
Z = clf.predict(X_test)
accuracy = clf.score(X_test,y_test)
print(accuracy)
```

# la variance dans le resultat dessous est de tres faible avec une valeur  de 0.00068 ce qui sinifie que le model ne varie pas beauucoup entre les differents tirages

```python
n_tirages = 100
accuracies = []

for _ in range (n_tirages):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9)
    model = tree.DecisionTreeClassifier()
    model.fit(X_train, y_train)
    
    accuracy = model.score(X_test,y_test)
    accuracies.append(accuracy)
    
mean_acuracy = np.mean(accuracies)
variance_accuracy = np.var(accuracies)

print(f"Accuracy moyenne :{mean_acuracy}")
print(f"Variance de l'Accuracy :{variance_accuracy}")
```

# Nous concluons que le modèle d'ensemble a une milleure stabilité vu la  valeur de la variance de l'accuracy qui est lus faible qu celle du modèle de base

```python
n_tirages = 100
accuracies2 = []

for _ in range (n_tirages):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9)
    model = BaggingClassifier(tree.DecisionTreeClassifier(),max_samples=0.5, max_features=0.5, n_estimators=200)

    model.fit(X_train, y_train)
    
    accuracy = model.score(X_test,y_test)
    accuracies2.append(accuracy)
    
mean_acuracy = np.mean(accuracies2)
variance_accuracy = np.var(accuracies2)

print(f"Accuracy moyenne :{mean_acuracy}")
print(f"Variance de l'Accuracy :{variance_accuracy}")
```

# L'accuracy augmente en fonction du nombre d'esimateurs pour atteindre son pic vers 75 estimateurs (arbres), puis se stabilise autour de cette valeur

```python

# Séparer les données en ensemble d'apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
# Liste des différents nombres d'estimateurs
n_estimators_range = range(1, 201, 5)
accuracies = []
# Boucle pour tester différents n_estimators
for n in n_estimators_range:
    # Créer et entraîner le modèle
    model = BaggingClassifier(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    # Calculer l'accuracy
    accuracy = model.score(X_test, y_test)
    accuracies.append(accuracy)
# Tracer le graphique
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, accuracies, marker='o')
plt.title('Accuracy vs. Nombre Estimateurs')
plt.xlabel('Nombre Estimateurs')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

la valeur optimale de max_features est de 0.5

# Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
digits = load_digits()
X, y = digits.data, digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.90)
# AdaBoost basé sur 200 arbres de décision
clf = AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=5),n_estimators=200, learning_rate=2)
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

# # Paramètres pour la variation de max_depth

Valeur opimale de max_depth = 7

```python

max_depth_range = range(1, 11)
learning_rate_fixed = 2  # Taux d'apprentissage constant
accuracies_depth = []
# Boucle pour tester différentes valeurs de max_depth
for max_depth in max_depth_range:
    clf = AdaBoostClassifier(
        tree.DecisionTreeClassifier(max_depth=max_depth),
        n_estimators=200,
        learning_rate=learning_rate_fixed
    )
    clf.fit(X_train, y_train)
    # Calculer l'accuracy
    accuracy = clf.score(X_test, y_test)
    accuracies_depth.append(accuracy)
# Tracer le graphique pour max_depth
plt.figure(figsize=(10, 6))
plt.plot(max_depth_range, accuracies_depth, marker='o')
plt.title(f'Accuracy vs. max_depth (learning_rate = {learning_rate_fixed})')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

# # Paramètres pour la variation de learning_rate

Valeur opimale de learning_rate = 1.7

```python

learning_rate_range = np.linspace(0.1, 3, 10)  # Valeurs de 0.1 à 3 avec un pas
max_depth_fixed = 5  # Profondeur maximale constante
accuracies_learning_rate = []
# Boucle pour tester différentes valeurs de learning_rate
for learning_rate in learning_rate_range:
    clf = AdaBoostClassifier(
        tree.DecisionTreeClassifier(max_depth=max_depth_fixed),
        n_estimators=200,
        learning_rate=learning_rate
    )
    clf.fit(X_train, y_train)
    # Calculer l'accuracy
    accuracy = clf.score(X_test, y_test)
    accuracies_learning_rate.append(accuracy)
# Tracer le graphique pour learning_rate
plt.figure(figsize=(10, 6))
plt.plot(learning_rate_range, accuracies_learning_rate, marker='o')
plt.title(f'Accuracy vs. learning_rate (max_depth = {max_depth_fixed})')
plt.xlabel('learning_rate')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

```
