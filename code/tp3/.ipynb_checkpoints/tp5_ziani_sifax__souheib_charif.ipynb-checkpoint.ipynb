```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
```

```python
# Chargement des données
iris = datasets.load_iris()
X, y = iris.data[:, :2], iris.target
# On conserve 50% du jeu de données pour l'évaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
C = 1.0 # paramètre de régularisation
lin_svc = svm.LinearSVC(C=C)
lin_svc.fit(X_train, y_train)
```

```python
# Calcul du score 
score = lin_svc.score(X_test, y_test)
print(f'Score d\'échantillons bien classifiés : {score:.2f}')
```

```python
# Créer la surface de décision discretisée
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
# Pour afficher la surface de décision on va discrétiser l'espace avec un
h = max((x_max - x_min) / 100, (y_max - y_min) / 100)
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
# Surface de décision
Z = lin_svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
# Afficher aussi les points d'apprentissage
plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k',
c=y_train, cmap=plt.cm.coolwarm)
plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test,
cmap=plt.cm.coolwarm)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title("LinearSVC")
```

```python
C_values = [0.01, 0.1, 1, 3, 5, 7, 9, 10, 100]

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

plt.figure(figsize=(15, 10))

for i, C in enumerate(C_values):
    lin_svc = svm.LinearSVC(C=C, random_state=42)
    lin_svc.fit(X_train, y_train)

    Z = lin_svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(3, 3, i + 1)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', marker='o', label='Entraînement')
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolors='k', marker='s', label='Test')
    plt.title(f'SVM avec C = {C}')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel('Attribut 1')
    plt.ylabel('Attribut 2')
    plt.legend()

plt.tight_layout()
plt.show()
```

D’après la visualisation ci-dessus, ce modèle n'est pas adapté au problème.

Etant donné que les les données ne sont pas linéairement séparables, utiliser un noyau non-linéaire (Geussien par exemple) peut améliorer les performances du modèle.

```python
lin_svc = svm.LinearSVC(C=C).fit(X_train, y_train)
svc = svm.SVC(kernel='linear', C=C).fit(X_train, y_train)
titles = ['SVC with linear kernel', 'LinearSVC (linear kernel)']
fig = plt.figure(figsize=(12, 4.5))
for i, clf in enumerate((svc, lin_svc)):
    plt.subplot(1, 2, i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    # Utiliser une palette de couleurs
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    # Afficher aussi les points d'apprentissage
    plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k', c=y_train, cmap=plt.cm.coolwarm)
    plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test, cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title(titles[i])
    plt.show()
```

 # Utilisation des quatre variables du jeu de données

```python
X, y = iris.data[:, :4], iris.target
# On conserve 50% du jeu de données pour l'évaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
C = 1
linear_svc = svm.LinearSVC(C=C, random_state=42, max_iter=10000)
linear_svc.fit(X_train, y_train)

score = linear_svc.score(X_test, y_test)
print(f'Score de classification avec les 4 attributs : {score:.2f}')
```

Oui le score de classification s'est améliorer nettement en utilisant les quatre variables du dataset (passer de 0.83 à 0.96).

L'utilisation de tous les attributs permet d'exploiter toutes les dimensions disponibles du jeu de données, ce qui peut améliorer la capacité discriminante du modèle par rapport à l'utilisation d'un sous-ensemble des attributs. De plus, il est plus facile de trouver un hyperplan séparateur dans des espaces de grande dimension.

# Jeu de données Digits

```python
from sklearn.model_selection import GridSearchCV # pour l'optimisation du paramètre C
from sklearn.metrics import accuracy_score
import time 

# Charger le jeu de données Digits
digits = datasets.load_digits()

# Extraire les features (X) et les labels (y)
X, y = digits.data, digits.target

# Diviser en jeu d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

# Créer un classifieur LinearSVC
svc = svm.LinearSVC(max_iter=10000)

# Définir les valeurs possibles de l'hyperparamètre C pour l'optimisation
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

# Utiliser GridSearchCV pour optimiser C
grid_search = GridSearchCV(svc, param_grid, cv=5)

# Mesurer le temps d'apprentissage
start_time = time.time()

grid_search.fit(X_train, y_train)

end_time = time.time()
elapsed_time = end_time - start_time

# Afficher la meilleure valeur de C et les performances
best_C = grid_search.best_params_['C']
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Meilleur hyperparamètre C : {best_C}")
print(f"Score de classification en test : {accuracy:.4f}")
print(f"Temps d'apprentissage : {elapsed_time:.2f} secondes")
```
