```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
```

```python
# Chargement des données
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

```python
# On conserve 50% du jeu de données pour l'évaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
X, y = iris.data[:,:2], iris.target
# On conserve 50% du jeu de données pour l'évaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
clf = svm.SVC(C=0.1, kernel='rbf', gamma=0.25)
clf.fit(X_train, y_train)
# Pour afficher la surface de décision on va discrétiser l'espace avec un 

h = .02
# Créer la surface de décision discretisée
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
# Surface de décision
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
# Afficher aussi les points d'apprentissage
plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k', 
c=y_train, cmap=plt.cm.coolwarm)
plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test, 
cmap=plt.cm.coolwarm)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title("SVM RBF")
```

# Testez l’effet du paramètre d’échelle du noyau (gamma)

```python
# Liste de valeurs de gamma à tester
gamma_values = [0.01, 0.1, 1, 10]
# Fixer une valeur constante pour C
C_fixed = 1.0
# Surface de décision discrétisée
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

for gamma in gamma_values:
    clf = svm.SVC(C=C_fixed, kernel='rbf', gamma=gamma)
    clf.fit(X_train, y_train)
    # Surface de décision
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Tracer la surface de décision
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    # Afficher les points d'apprentissage et de test
    plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k', c=y_train, cmap=plt.cm.coolwarm)
    plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test, cmap=plt.cm.coolwarm)
    # Ajouter le titre avec la valeur de gamma
    plt.title(f"SVM RBF avec gamma={gamma}, C={C_fixed}")
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.legend()
    plt.show()
```

# Testez l’effet du paramètre de régularisation C

```python
# Liste de valeurs de C à tester
C_values = [0.01, 0.1, 1, 10]
# Fixer une valeur constante pour gamma
gamma_fixed = 0.25
# Boucle pour tester différentes valeurs de C
for C in C_values:
    clf = svm.SVC(C=C, kernel='rbf', gamma=gamma_fixed)
    clf.fit(X_train, y_train)
    # Surface de décision
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Tracer la surface de décision
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    # Afficher les points d'apprentissage et de test
    plt.scatter(X_train[:, 0], X_train[:, 1], label="train", edgecolors='k', c=y_train, cmap=plt.cm.coolwarm)
    plt.scatter(X_test[:, 0], X_test[:, 1], label="test", marker='*', c=y_test, cmap=plt.cm.coolwarm)
    # Ajouter le titre avec la valeur de C
    plt.title(f"SVM RBF avec C={C}, gamma={gamma_fixed}")
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.legend()
    plt.show()
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
X, y = digits.data, digits.target
print(X.shape)
print(y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# SVM avec un noyau linéaire sur les données Digits

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix


# Créer un modèle SVM avec un noyau linéaire
clf = svm.SVC(kernel='linear')
# Entraîner le modèle
clf.fit(X_train, y_train)
# Prédictions sur le jeu de test
y_pred = clf.predict(X_test)
# Afficher la précision et le rapport de classification
print("Classification report:\n", classification_report(y_test, y_pred))
# Afficher la matrice de confusion
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)
# Optionnel : Afficher la matrice de confusion sous forme graphique
plt.matshow(cm, cmap=plt.cm.Blues)
plt.title('Matrice de confusion')
plt.colorbar()
plt.ylabel('Classe réelle')
plt.xlabel('Classe prédite')
plt.show()


```

# SVM avec un noyau gaussien sur les données Digits

```python
# Créer un modèle SVM avec un noyau RBF (gaussien)
clf = svm.SVC(kernel='rbf', gamma=0.001, C=1.0)
# Entraîner le modèle
clf.fit(X_train, y_train)
# Prédictions sur le jeu de test
y_pred = clf.predict(X_test)
# Afficher la précision et le rapport de classification
print("Classification report:\n", classification_report(y_test, y_pred))
# Afficher la matrice de confusion
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)
# Optionnel : Afficher la matrice de confusion sous forme graphique
plt.matshow(cm, cmap=plt.cm.Blues)
plt.title('Matrice de confusion')
plt.colorbar()
plt.ylabel('Classe réelle')
plt.xlabel('Classe prédite')
plt.show()
```
